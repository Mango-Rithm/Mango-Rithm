#### <문제> 사칙연산 - 백준 10869번

#### <문제> 나머지 - 백준 10430번  
(A+B)%C는 ((A%C) + (B%C))%C 와 같을까?

(A×B)%C는 ((A%C) × (B%C))%C 와 같을까?

세 수 A, B, C가 주어졌을 때, 위의 네 가지 값을 구하는 프로그램을 작성하시오.

````python
a, b, c = map(int, input().split())
print((a + b) % c)
print(((a % c) + (b % c)) % c)
print((a * b) % c)
print(((a % c) * (b % c)) % c)
````

(1)과 (2)위치에 들어갈 세 자리 자연수가 주어질 때 (3), (4), (5), (6)위치에 들어갈 값을 구하는 프로그램을 작성하시오.

````python
a = int(input())
b = input()
b1 = int(b[2])
b2 = int(b[1])
b3 = int(b[0])
b = int(b)

print(a * b1)
print(a * b2)
print(a * b3)
print(a * b)
````

````python
# 더 나은 코드
a = int(input())
b = input()

for i in range(2, -1, -1):
    print(a * int(b[i]))
    i -= 1
print(a * int(b))
````
